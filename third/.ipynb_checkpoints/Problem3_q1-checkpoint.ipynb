Use TensorFlow with three layers of neural network, similar to Problem statement#2, to classify the data set in Problem statement#2. The first layer (input layer) has four neurons, the second layer has five neurons, and the third layer (output layer) has only one neuron. Try the same experience with the same number of neurons you used in each layer. You should use at least two different activation function and two different optimization functions. Plot your cost function.

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
import math
from IPython import display
import time


#input dataset
x1_list = []
x2_list = []
x3_list = []
x11_list = []
x22_list = []
x33_list = []
t_list = []


#m=int(input("please enter number of first layer (m) nodes: "))
#n=int(input("please enter number of second layer (n) nodes: "))

#Generating first data set
for i in range(50):
    x1 = random.uniform(-0.5,0.5)
    x2 = random.uniform(-0.5,0.5)
    x1_list.append(x1)
    x2_list.append(x2)
    x3_list.append(1)
    #taking target as 0
    t_list.append(0)

#Generating second data set
for i in range(25):    
    r = random.uniform(9,16)
    x_1 = random.uniform(-4,4)
    temp = r-pow(x_1,2)
    if(temp > 0):
        x_2 = math.sqrt(temp)
    x11_list.append(x_1)
    x22_list.append(x_2)
    t_list.append(1)
    x33_list.append(1)
    
for i in range(25):   
    r = random.uniform(9,16)
    x_1 = random.uniform(-4,4)
    temp = r-pow(x_1,2)
    if(temp > 0):
        x_2_temp = math.sqrt(temp)
        #taking negative input values
        x_2 = -(x_2_temp)
    x11_list.append(x_1)
    x22_list.append(x_2)
    t_list.append(1)
    x33_list.append(1)

#Appending input datasets
x_temp = [x1_list + x11_list , x2_list + x22_list , x3_list + x33_list] 
print(t_list)


# print(np.shape(x_temp))

#Transposing matrix to 100*3
x = np.array(x_temp).T
t=np.array(t_list)

#print(type(x))
'''
m=np.array([x1_list + x11_list])
n=np.array([x2_list + x22_list])
p=np.array([x3_list + x33_list])


print("m",type(m))
print("n",type(n))
print("p",type(p))
'''
```

```python
model = tf.keras.Sequential([
  keras.layers.Dense(units=4, input_shape=[3], activation=tf.nn.relu),
keras.layers.Dense(units=5, activation=tf.nn.relu),
keras.layers.Dense(units=1, activation=tf.nn.sigmoid)])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(x,t, epochs=100)
print(history)
```

```python
print(model.predict(x))
```

```python
import matplotlib.pyplot as plt

loss = history.history['loss']


epochs = range(len(loss))

plt.plot(epochs, loss, 'r', label='Error')

plt.title('Loss Curve')
plt.legend(loc=0)
plt.figure()


plt.show()
```

```python
model = tf.keras.Sequential([
  keras.layers.Dense(units=4, input_shape=[3], activation=tf.nn.elu),
keras.layers.Dense(units=5, activation=tf.nn.elu),
keras.layers.Dense(units=1, activation=tf.nn.sigmoid)])

model.compile(optimizer='RMSprop', loss='mean_squared_error')
```

```python
history = model.fit(x,t, epochs=100)
print(history)
```

```python
import matplotlib.pyplot as plt

loss = history.history['loss']


epochs = range(len(loss))

plt.plot(epochs, loss, 'r', label='Error')

plt.title('Loss Curve')
plt.legend(loc=0)
plt.figure()


plt.show()
```
