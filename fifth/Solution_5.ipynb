```python
training_folder='D:\\animal_feature\\train'
validation_folder='D:\\animal_feature\\validation'
```

```python
import tensorflow as tf
#importing image data generator
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.layers.normalization import BatchNormalization

#initializing batch size to be 50
batchsize=50

#rescaling the image to the values between 0 and 1
train_imagegen = ImageDataGenerator( rescale = 1/255)
validation_Imagegen = ImageDataGenerator( rescale = 1/255)


train_generator = train_imagegen.flow_from_directory(training_folder, 
                                                    batch_size=batchsize,
                                                    class_mode='categorical',
                                                    target_size=(150, 150))

validation_generator = validation_Imagegen.flow_from_directory(validation_folder, 
                                                        batch_size=batchsize,
                                                        class_mode ='categorical',
                                                        target_size = (150, 150))

#print(train_imagegen)

#defining model as Sequential
model = tf.keras.models.Sequential([
 # Note the input shape is the desired size of the image 150x150 with 3 bytes color
 #first covolutional layer
 tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
 tf.keras.layers.MaxPooling2D(2,2),
 #second covolutional layer
 tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),
 #third covolutional layer  
 tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),
 #fourth covolutional layer
 tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),
 #adding dropout layer
 tf.keras.layers.Dropout(0.2),
 #fourth covolutional layer
 tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),
 #adding dropout layer
 tf.keras.layers.Dropout(0.2),
 # Flatten the results to feed into a DeepNeuralNetwork
 tf.keras.layers.Flatten(),
 # Dense layer with 512 neurons which are fully connected
 tf.keras.layers.Dense(512, activation='relu'),    
 #tf.keras.layers.Dense(512, activation='relu', kernel_initializer='he_uniform', kernel_regularizer=regularizers.l2(0.001)),
 # 12 output neurons. It will contain a value from 0-11
 tf.keras.layers.Dense(12, activation='softmax')
])

#model compilation
model.compile(tf.keras.optimizers.Adam(lr=0.001),  loss='categorical_crossentropy', metrics = ['acc'])

#fitting model
history = model.fit(train_generator,steps_per_epoch=150,epochs=30,validation_data=validation_generator,validation_steps=50)
```

```python
#storing all the image paths, names and features to its respective variables
import numpy as np
from keras.preprocessing import image
import os
folder_path='D:\\animal_feature\\train'
images = []
feature_list=[]
img_path_list=[]
img_name=[]

for img in os.listdir(folder_path):
    #print(img) #cats/dogs
    for img1 in os.listdir(folder_path+'/'+img):
        #print("img1",img1) #image name cat.1265.jpg
        imagefolder_path=os.path.join(folder_path, img)
        #print(imagefolder_path)
        img_name.append(img1)
        image_path=os.path.join(imagefolder_path, img1)
        #print(image_path)
        x=image.load_img(image_path, target_size=(150, 150))
        x=image.img_to_array(x)
        x=np.expand_dims(x, axis=0)
        images = np.vstack([x])
        #print(x)
        features = model_reduced.predict(images)
        #print(classes[0])
        feature_list.append(features)
        #print(features)
        img_name.append(img1)
        img_path_list.append(image_path)
```

```python
#calculating the euclidean distance and get the minimum distance from it
from scipy.spatial import distance
euclid_feature=[]
for i in feature_list:
        dist=distance.euclidean(i, face_features)
        euclid_feature.append(dist)
min_value=min(euclid_feature)
min_index = euclid_feature.index(min_value)
print("minimum distance",min_value)
print("index of the minimum distance:" ,min_index)

#printing path
feature_path=img_path_list[min_index]
print(feature_path)
```
